### Techniques to create or use dummy dataframes

- To test your code
- To start with some data for more experimentation
- Interviewers might not be interested to use a dataframe from csv file

1. `Pandas` util.testing module dataframes
  - might get deprecated shortly. but handy option
  - you need to select the dataset based on your use  case. 

2. `scikit-learn` dataframes
 - know how to convert the values after loading the dataframes into a pandas dataframe
 - you need to select the dataset based on your use  case. 
3. `numpy` to create some toy/dummy dataframe from scratch
  - always going to be numeric

```python
# Technique 1
pd.util.testing.makeDataFrame()
```

```python
pd.util.testing.makeMissingDataframe()
```

```python
pd.util.testing.makeMixedDataFrame()
```
