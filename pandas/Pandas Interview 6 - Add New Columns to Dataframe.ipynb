### How to add one or more columns in your dataframe

1. how to add one column in your dataframe
2. how to add more than one columns in your dataframe

```python
import numpy as np
```

```python
array = np.random.rand(5,5)
df = pd.DataFrame(array)
df
```

```python
df['firstway_a'] = -100
```

```python
df
```

```python
df['firstway_b'] = -200
```

```python
df
```

```python
mylist = [i for i in range(100, 600, 100)]
mylist
```

```python
df['mylist'] = mylist
```

```python
df
```

```python
df[['secondway_a', 'secondway_b']] = -300, -400

df
```

```python
df['thirdway_a'] = df['firstway_a'] * 10
df
```

```python
df['thirdway_b'] = df[0] * 100 + df['thirdway_a'] + df['secondway_b']
df
```

```python
df[['thirdway_a', 'thirdway_b']] = df['firstway_a'] * 10, df[0] * 100 + df['thirdway_a'] + df['secondway_b']
```

```python
df = df.assign(thirdway_a = df['firstway_a'] * 10,
               thirdway_b = df[0] * 100 + df['thirdway_a'] + df['secondway_b'])
```
