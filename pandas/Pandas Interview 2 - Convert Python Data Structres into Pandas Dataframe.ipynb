```python
# Import Library
import pandas as pd
```

```python
# First list, contains hypothetical ids
idList = [10, 20, 30, 40, 50]
```

```python
# Dataframe from list
pd.DataFrame(idList)
```

```python
# Dataframe from a list with column name
pd.DataFrame(idList, columns=['id'])
```

```python
# Second list, containing hypothetical names of corresponding ids
nameList = ['A', 'B', 'C', 'D', 'E']
```

```python
# Sneak peek
idList, nameList
```

```python
# We created a dataframe with one column from one list
# If we have two lists and want to create a dataframe with two
# columns from them, can we use the technique we just saw?
lists = [idList, nameList]
pd.DataFrame(lists, columns=['id', 'name'])
```

```python
# One remedy is, if you have column values in lists, zip them up
zippedList = zip(idList, nameList)
```

```python
# And then use that zip object to create a dataframe
pd.DataFrame(zippedList, columns=['id', 'name'])
```

```python
# We can also create each row in our dataframe as lists and put these lists
# in another list
listOfLists = [
    [1, 'A'],
    [2, 'B'],
    [3, 'C']
]
pd.DataFrame(listOfLists, columns=['id', 'name'])
```

```python
# Tuple and list operates in a similar way
idTuple= (10,20,30,40,50)
pd.DataFrame(idTuple, columns=['id'])
```

```python
# Same error as we got before
tuples = (
    (10,20,30,40,50), 
     ('A', 'B', 'C', 'D', 'E')
)
pd.DataFrame(tuples, columns=['id', 'name'])
```

```python
# Zipping up tuples can be a solution
nameTuple = ('A', 'B', 'C', 'D', 'E')
zippedTuple = zip(idTuple, nameTuple)
```

```python
pd.DataFrame(zippedTuple, columns=['id', 'name'])
```

```python
# We also can create a tuple of values for each row and put them in a tuple
tupleOfTuples = (
    (10, 'A'), 
    (20, 'B'), 
    (30, 'C'),
    (40, 'E'),
    (50, 'E')
)
pd.DataFrame(tupleOfTuples, columns=['id', 'name'])
```

```python
# A dictionary can be created where the keys are the column names and values 
# are the row values of our dataframe in a list
myDict = {
    'id': idList,
    'name': nameList
}
pd.DataFrame(myDict)
```

```python
# We can as well create a dictionary for each row in a dataframe and keep
# them in a list...
listOfDicts = [
    {'id': 10, 'name': 'A'},
    {'id': 20, 'name': 'B'}
]

pd.DataFrame(listOfDicts)
```

```python
# ... or in a tuple
tupleOfDicts = (
    {'id': 10, 'name': 'A'},
    {'id': 20, 'name': 'B'}
)

pd.DataFrame(tupleOfDicts)
```
