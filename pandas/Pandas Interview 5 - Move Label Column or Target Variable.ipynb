### How to move label column or target variable to a desired location/ position

1. Create a dataset or dataframe from scratch
2. Add labels or target variables to that dataset
3. Pop out the column which is label or target variable
4. Insert that popped out column into a desired location

```python
import numpy as np
array = np.random.rand(10,10)
```

```python
df = pd.DataFrame(array,
                  columns=list('abcdefghij'))
```

```python
df.head(10)
```

```python
df['label'] = np.random.randint(2, size=10)
```

```python
df.head(10)
```

```python
label = df.pop('label')
```

```python
df
```

```python
df.insert(0, 'label', label)
```
