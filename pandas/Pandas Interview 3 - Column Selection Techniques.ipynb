```python
import pandas as pd
from sklearn import datasets
```

```python
dataset = datasets.load_diabetes()
```

```python
df = pd.DataFrame(data=dataset.data,
                  columns=dataset.feature_names)
df.head()
```

```python
cols_to_keep = ['age', 'sex', 'bmi', 'bp']
df[cols_to_keep]
```

```python
df.iloc[:, :4]
```

```python
df.iloc[:, [0, 1, 2, 3]]
```

```python
df.loc[:, cols_to_keep]
```

```python
df.loc[:, ::2]
```

```python
df.columns
```

```python
df.drop(columns=[i for i in df.columns if i not in cols_to_keep],
        axis=1)
```

```python
cols_to_keep
```

```python
df.take([0, 1, 2, 3], axis=1)
```

```python
df.filter(cols_to_keep, axis=1)
```

```python
df.get(cols_to_keep)
```

```python
df.sort_index(axis=1)
```

```python
df.sort_index(axis=1).truncate(before='s1',
                               after='s6',
                               axis=1)
```
