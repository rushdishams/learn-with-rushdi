```python
import pandas as pd
```

### Difference between `loc` and `iloc

1. `loc` gets rows (and/or columns) with particular labels
2. `iloc` gets rows (and/ or columns) with particular integer locations

```python
# Create a dummy dataframe
df = pd.DataFrame(
    {
        'id': [10, 20, 30, 40, 50],
        'name': ['Alice', 'Bob', 'Charlie', 'David', 'Elon']
    },
  index=[200, 0, 100, 2, 300] 
)
df.head()
```

```python
# Single value for loc
df.loc[100]
```

```python
# Single value for iloc
df.iloc[100]
```

```python
df.head()
```

```python
# Slice in loc
df.loc[200:100]
```

```python
# Slice in iloc
df.iloc[200:100]
```

```python
df.head()
```

```python
# Slice in loc
df.loc[0:200]
```

```python
# List in loc
df.loc[[200, 300, 2]]
```

```python
# List in iloc
df.iloc[[0, 4, 2]]
```

```python
df.head()
```

```python
# Row and column in loc
df.loc[[200, 300, 2], 'name']
```

```python
# Row and column in iloc
df.iloc[[0, 4, 2], 'name']
```

```python
# Row and column in iloc
df.iloc[[0, 4, 2], 0:1]
```

```python
# Row and column in iloc
df.iloc[[0, 4, 2], 1]
```

```python
# Row and column in iloc
df.iloc[[0, 4, 2], [0, 1]]
```

```python
# Row and column in iloc
df.iloc[[0, 4, 2], 0:3]
```

```python
# Full dataframe with loc
df.loc[:, :]
```

```python
# Full dataframe with iloc
df.iloc[:, :]
```

```python
# Add another column
df['salary'] = [1000, 2000, 3000, 4000, 5000]
df.head()
```

```python
# Condition in loc
df.loc[df['salary'] > 2500]
```

```python
# Multiple conditions in loc
df.loc[(df['salary'] > 2500) &
       (~df['name'].isin(['Charlie']))]
```

```python
df.head()
```

```python
# Resetting index creates a new column with old index
df.reset_index()
```

```python
# Dropping old index as column after resetting index
df = df.reset_index(drop=True)
df.head()
```

```python
# slicing for loc in inclusive
df.loc[0:3]
```

```python
# slicing for iloc is exclusive
df.iloc[0:3]
```
